```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("./data/shelter_origin.csv", encoding="utf-8")
df_original = df.copy()

print(df.columns)
len(df.columns)
print("삭제 전 행 개수:", len(df),"삭제 전 열의 개수:",len(df.columns))
```

```python
# 운영상태가 '사용중지'인 행 제거
df = df[df["운영상태"] != "사용중지"]
df = df[df["운영상태"] != "일시정지"]
print("삭제 후 행 개수:", len(df))
```

```python
df = df.drop(columns=["번호","지정일자","소재지전체주소","데이터갱신구분","위도(도)", "위도(분)", "위도(초)", 
               "경도(도)", "경도(분)", "경도(초)","좌표정보X(EPSG5179)", "좌표정보Y(EPSG5179)","해제일자"])
print(df.columns)
len(df.columns)
```

```python
df = df.dropna(subset=['도로명우편번호'])
```
